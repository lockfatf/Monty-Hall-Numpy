```python
print("En changeant de porte, le joueur a gagné {} sur 10000 parties."
      .format(sum(play.play_game(st.Strategie.CHANGER, 10000))))
      
print("En gardant son choix initial, le joueur a gagné {} sur 10000 parties."
      .format(sum(play.play_game(st.Strategie.GARDER, 10000))))
```

```python
plot = plt.scatter(range(10000), play.play_game(st.Strategie.CHANGER, 10000))
```

```python
plot = plt.bar([1,2],[float(sum(play.play_game(st.Strategie.CHANGER, 10000))), 
               float(sum(play.play_game(st.Strategie.GARDER, 10000)))], 
        tick_label=["Changer","Garder"])
```

```python
gains_changer = []
gains_garder = []
samples =  [1000, 10000, 20000, 50000, 80000, 100000]
for tours in samples:
    gains_changer.append(play.play_game(st.Strategie.CHANGER, tours))
    gains_garder.append(play.play_game(st.Strategie.GARDER, tours))
```

```python
figure = plt.figure()
plot = plt.scatter(samples, [sum(x) for x in gains_changer])
plot = plt.scatter(samples, [sum(x) for x in gains_garder])
```
